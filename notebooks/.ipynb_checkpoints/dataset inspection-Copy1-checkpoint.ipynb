```python
#create function to check if force was used
def force(c):
  if c['pf_hands'] == "Y":
    return 1
  elif c['pf_wall'] == "Y":
    return 1
  elif c['pf_grnd'] == "Y":
    return 1
  elif c['pf_drwep'] == "Y":
    return 1
  elif c['pf_ptwep'] == "Y":
    return 1
  elif c['pf_wall'] == "Y":
    return 1
  elif c['pf_baton'] == "Y":
    return 1
  elif c['pf_hcuff'] == "Y":
    return 1
  elif c['pf_pepsp'] == "Y":
    return 1
  elif c['pf_other'] == "Y":
    return 1
  else:
    return 0
```

```python
#creates new column based on query above
gdf['force'] = gdf.apply(force, axis=1)
```

```python
# create function to determine if subject arrested or summoned
def arrest(c):
  if c['arstmade'] == "Y":
    return 1
  elif c['sumissue'] == "Y":
    return 1
  else:
    return 0
```

```python
#creates new column based on query above
gdf['arrest'] = gdf.apply(arrest, axis=1)
```

```python
gdf.age = pd.to_numeric(gdf.age, errors='coerce').fillna(0).astype(np.int64)
```

```python
# determine if minor
def minor(c):
  if c['age'] < 18:
    return 1
  else:
    return 0
```

```python
gdf['minor'] = gdf.apply(minor, axis=1)
```

```python
def weapon(c):
    if c['pistol'] == "Y":
        return 1
    elif c['knifcuti'] == "Y":
        return 1
    else:
        return 0
```

```python
gdf['weapon'] = gdf.apply(weapon, axis=1)
```
